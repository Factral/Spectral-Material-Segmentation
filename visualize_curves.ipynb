```python
import glob
import numpy as np
import matplotlib.pyplot as plt

materials_name = glob.glob('materials_numpy/*.npy')
materials = np.array([np.load(m) for m in materials_name])

for material, name in zip(materials, materials_name):
    if name == 'materials_numpy\water.npy':
        continue
    print(material, name)
    plt.plot(material)
```
